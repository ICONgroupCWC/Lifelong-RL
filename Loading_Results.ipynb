```python
import numpy as np

Reward_episoeds = np.load('reward_L2RL.npy')
episodes                = 1000

import matplotlib.pyplot as plt
time = np.arange(episodes)
plt.figure(figsize=(10, 6))
plt.plot(time, Reward_episoeds, label='Reward ETA = 0.2', color='b', marker='*', linestyle='-', linewidth=3, markersize=4, markeredgecolor='black')
plt.xlabel('Slots')
plt.ylabel('reward')
plt.title('reward, slots')
plt.grid(True)
#plt.savefig('TX1 Data buffer lenreward_L2RLZgth vs Time.png')  # Saves the figure as a PNG file
plt.show()
```

```python
window_size = 10
smoothed_rewards = np.convolve(Reward_episoeds, np.ones(window_size)/window_size, mode='valid')

# Adjusting the time array to match the length of the smoothed_rewards
time = np.arange(len(smoothed_rewards))

plt.figure(figsize=(10, 6))
plt.plot(time, smoothed_rewards, label='Smoothed Reward', color='b', marker='*', linestyle='-', linewidth=3, markersize=4, markeredgecolor='black')
plt.xlabel('Slots')
plt.ylabel('Reward')
plt.title('Smoothed Reward vs Slots')
plt.grid(True)
plt.legend()
plt.show()
```
